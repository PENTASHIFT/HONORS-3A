```python
import string

import pandas as pd
from nltk import word_tokenize
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

stop_words.add("im")
string.punctuation += 'â€™'
```

```python
schizo_csv = pd.read_csv("../response_data/schizo_responses.csv",
                      lineterminator='\n')
control_csv = pd.read_csv("../response_data/control_responses.csv",
                         lineterminator='\n')

schizo_responses = schizo_csv.sample(n=2500)["tweet"].tolist()
control_responses = control_csv.sample(n=2500)["tweet"].tolist()

schizo_responses = [tweet.translate(str.maketrans('', '', string.punctuation)) for tweet in schizo_responses]
control_responses = [tweet.translate(str.maketrans('', '', string.punctuation)) for tweet in control_responses]

schizo_responses_tokenized = [word_tokenize(tweet)[2:] for tweet in schizo_responses]
control_responses_tokenized = [word_tokenize(tweet)[2:] for tweet in control_responses]

schizo_responses_lowered = [[w.lower() for w in tweet] for tweet in schizo_responses_tokenized]
control_responses_lowered = [[w.lower() for w in tweet] for tweet in control_responses_tokenized]

schizo_responses_filtered = [[w for w in tweet if w not in stop_words] for tweet in schizo_responses_lowered]
control_responses_filtered = [[w for w in tweet if w not in stop_words] for tweet in control_responses_lowered]
```

### Brown Clustering

```python
import brownclustering as bc
```

```python
schizo_corpus = bc.Corpus(schizo_responses_lowered, 0.001)
control_corpus = bc.Corpus(control_responses_lowered, 0.001)

schizo_clustering = bc.BrownClustering(schizo_corpus, 6)
control_clustering = bc.BrownClustering(control_corpus, 6)

schizo_clustering.train()
control_clustering.train()
```

```python
print(schizo_clustering.get_similar("think"))
print(control_clustering.get_similar("think"))
```

### LDA Clustering

```python
import gensim
import gensim.corpora as corpora
import pyLDAvis.gensim
```

```python
schizo_id2word = corpora.Dictionary(schizo_responses_filtered)
control_id2word = corpora.Dictionary(control_responses_filtered)

schizo_corpus = [schizo_id2word.doc2bow(text) for text in schizo_responses_filtered]
control_corpus = [control_id2word.doc2bow(text) for text in control_responses_filtered]
```

```python
num_topics = 10
schizo_lda_model = gensim.models.LdaMulticore(corpus=schizo_corpus,
                                                id2word=schizo_id2word,
                                                num_topics=num_topics
                                             )
control_lda_model = gensim.models.LdaMulticore(corpus=control_corpus,
                                                id2word=control_id2word,
                                                num_topics=num_topics
                                              )
```

```python
vis = pyLDAvis.gensim.prepare(topic_model=schizo_lda_model, corpus=schizo_corpus, dictionary=schizo_id2word)
pyLDAvis.display(vis)
```

```python
vis = pyLDAvis.gensim.prepare(topic_model=control_lda_model, corpus=control_corpus, dictionary=control_id2word)
pyLDAvis.display(vis)
```
